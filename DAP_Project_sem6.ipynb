**Perform:**

---



**Simple Linear Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Independent variable (X)
X = df[['device_hours_per_day']]

# Dependent variable (y)
y = df['happiness_score']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Output
print("Predicted Values:", y_pred)
```

Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Evaluation
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Model Evaluation:")
print("MSE:", mse)
print("MAE:", mae)
print("R2 Score:", r2)
```

Visualization of Simple Linear Regression

```python
import matplotlib.pyplot as plt

# Scatter plot (Actual vs Predicted)
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Happiness Score")
plt.ylabel("Predicted Happiness Score")
plt.title("Actual vs Predicted Happiness Score")

# Perfect prediction line
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])

plt.show()
```

**Analyze:**

---

1. Covariance
2. Correlation

**Covariance**

```python
cov_val = df['device_hours_per_day'].cov(df['happiness_score'])
print("Covariance between Device Hours & Happiness Score:", cov_val)
```

The covariance value of -3.61 indicates a negative relationship between device usage and happiness score.

This means that as device hours increase, the happiness score tends to decrease. The negative sign shows an inverse relationship between the two variables.

Unlike correlation, covariance does not have a fixed range, so it is mainly used to understand the direction of the relationship, not the strength.

**Correlation**

```python
# Correlation between device hours and happiness
corr_value = df['device_hours_per_day'].corr(df['happiness_score'])

print("Correlation (Device Hours vs Happiness Score):", corr_value)
```

Device hours per day shows a negative correlation with happiness score, indicating that higher screen time may reduce happiness.

# **Task 10: Classification Task**

---



---



**Convert the problem into a classification problem**

---



 => Convert the problem into a classification problem

In this task, the problem is converted into a classification problem.
The target variable used is 'high_risk_flag'.

0 = Not High Risk
1 = High Risk

Since the output has two classes, this is a binary classification problem.

```python
print("Target Variable: high_risk_flag")
print("Classes:")
print("0 = Not High Risk")
print("1 = High Risk")

# Check class distribution
df['high_risk_flag'].value_counts()
```

---

---

**Accuracy**

---



```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

**Confusion Matrix**

---



```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```
